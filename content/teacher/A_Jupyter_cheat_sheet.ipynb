# Appendix A: Jupyter notebook cheat sheet

[Jupyter](https://jupyter.org) is a coding environment, based on **notebooks**. 



______
## 1) What is a Jupyter notebook? 📓

A notebook is a document that contains typically **executable code** and **readable text explanations**.

It is organized in **cells**. You can determine the extent of a cell by clicking on it (once). 
Cells can be of two different types:
- **Code** : usually Python, which also display the code output when executed
- **Text** : formatted in [markdown](https://www.markdownguide.org/getting-started/), which can also contain figures and equations.

____
## 2) Write and execute Python code 🐍

We can start writing Python code in a code-type cell:

```python
# this is a Python code cell
print("Hello world!")
```

The above cell can be selected by **clicking on it once**, and press:
- `ctrl` + `enter` to execute it
- `shift` + `enter` to execute it and select the next cell.

For instance, if we first execute:

we can access `a` in another cell:

```python
print(myVariable)
```

We can therefore define functions, import modules... The key is to execute the cells in **order**! It is possible to re-execute the cells as many times as desired.

**Tips**:
- Re-execute the cells during development for debugging
- To make sure the notebook is runnable, reset the kernel and re-execute all the cells in order by pressing the double-triangle button on the top bar
